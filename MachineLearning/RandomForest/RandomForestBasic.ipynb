```python
import pandas as pd
import datetime
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
```

```python
df=pd.read_csv("ks-projects-201801.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
#募集日数
df["deadline"]=pd.to_datetime(df["deadline"])
df["launched"]=pd.to_datetime(df["launched"])
df["days"]=(df["deadline"]-df["launched"]).dt.days
```

```python
#successfulとfailedのデータを抽出し、successful→1,failed→0
df=df.query("state == ['successful','failed' ]")
df["state"]=df["state"].replace("failed",0)
df["state"]=df["state"].replace("successful",1)
```

```python
#不要な行を削除
df = df.drop(["ID","name","deadline","launched","backers","pledged","usd pledged","usd_pledged_real","usd_goal_real"], axis=1)
df.head()
```

```python
#カテゴリ変数処理
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
#データ分割
train_data=df.drop("state",axis=1)
y=df["state"].values
x=train_data.values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1234)
```

```python
#ランダムフォレスト
clf=RandomForestClassifier(random_state=1234)
clf.fit(x_train,y_train)
print("socre=",clf.score(x_test,y_test))
```
